```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xTrain = np.array([1.0,2.0])
yTrain = np.array([300.0,500.0])
print("xTrain: ",xTrain)
print("yTrain: ",yTrain)
```

```python
print("xTrain.shape: ",xTrain.shape)
m = xTrain.shape[0]
print("m: ",m)
m = len(xTrain)
print("m: ",m)
m = xTrain.size
print("m: ",m)
```

```python
plt.scatter(xTrain,yTrain,color='r')
plt.title("Training Data")
plt.xlabel("price")
plt.ylabel("size")
plt.show()
```

```python
def computeModel(x,w,b):
    m = len(x)
    y = np.zeros(m)
    for i in range(m):
        y[i] = w*x[i] + b
    return y
```

```python
w = 200.0
b = 100.0
tempY = computeModel(xTrain,w,b)
plt.scatter(xTrain,yTrain,color='r',label='Training Data')
plt.plot(xTrain,tempY,color='b',label='Model')
plt.title("Training Data")
plt.xlabel("price")
plt.ylabel("size")
plt.legend()
plt.show()
```
